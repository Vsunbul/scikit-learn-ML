 ## KNeighborsClassifier

```python
# Import the pandas library, which is used for data manipulation and analysis.
import pandas as pd

# Import the numpy library, which is used for numerical computations.
import numpy as np

# Import the seaborn library, which is used for statistical data visualization.
import seaborn as sns
```

```python
# Read and assign data
df = pd.read_csv("weight-height.csv")
```

```python
# Data visualization with scatter plot

figure = sns.scatterplot(data = df, x = "Weight", y = "Height", hue = "Gender")
```

```python
# Overview of the data.

df.head()
```

```python
# Check for null (missing) data in the DataFrame 'df'

df.isnull().sum()
```

```python
#  Dividing independed and depended variable.

X = df.drop("Gender", axis = 1)

y = df["Gender"]
```

```python
# Import the KNeighborsClassifier from the scikit-learn library.
from sklearn.neighbors import KNeighborsClassifier

# Import the train_test_split function from the scikit-learn library.
from sklearn.model_selection import train_test_split
```

```python
# Split the data into training and testing sets, with 20% of the data used for testing and stratified sampling based on 'y'.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)

# Initialize the k-nearest neighbors classifier with 3 neighbors.
knn = KNeighborsClassifier(n_neighbors=3)

# Fit the classifier to the training data.
knn.fit(X_train, y_train)

# Calculate the accuracy of the classifier on the test data.
knn.score(X_test, y_test)
```

```python
knn.predict([[50,160]])
```
